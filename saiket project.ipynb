#Tasks 3: Customer Segmentation

```python
# Inspect the data
print(data1.head())
print(data1.dtypes)
```

```python
''' Create Segments
Segment by Tenure
You can create segments by grouping tenure into bins. For example, you can define the following segments:

0-12 months
13-24 months
25-36 months
37-48 months
49+ months'''
```

```python
# Create tenure segments
bins = [0, 12, 24, 36, 48, data1['tenure'].max()]
labels = ['0-12 months', '13-24 months', '25-36 months', '37-48 months', '49+ months']
data1['TenureSegment'] = pd.cut(data1['tenure'], bins=bins, labels=labels, right=True)
```

```python
#Segment by Monthly Charges
```

```python
# Create monthly charges segments
bins = [0, 30, 60, data1['MonthlyCharges'].max()]
labels = ['Low', 'Medium', 'High']
data1['ChargesSegment'] = pd.cut(data1['MonthlyCharges'], bins=bins, labels=labels, right=True)
```

```python
#Segment by Contract Type
# Ensure 'ContractType' is treated as a categorical variable
data1['ContractType'] = data1['Contract'].astype('category')
```

#Analyze Churn Rates Across Segments

```python
# Churn rates by tenure segment
tenure_churn_rate = data1.groupby('TenureSegment')['Churn'].mean()

# Churn rates by charges segment
charges_churn_rate = data1.groupby('ChargesSegment')['Churn'].mean()

# Churn rates by contract type
contract_churn_rate = data1.groupby('ContractType')['Churn'].mean()

print('Churn Rate by Tenure Segment:')
print(tenure_churn_rate)

print('\nChurn Rate by Monthly Charges Segment:')
print(charges_churn_rate)

print('\nChurn Rate by Contract Type:')
print(contract_churn_rate)
```

```python
#Visualize Churn Rates

import matplotlib.pyplot as plt
import seaborn as sns

# Create subplots for visualization
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

# Churn Rate by Tenure Segment
sns.barplot(x=tenure_churn_rate.index, y=tenure_churn_rate.values, ax=axes[0])
axes[0].set_title('Churn Rate by Tenure Segment')
axes[0].set_xlabel('Tenure Segment')
axes[0].set_ylabel('Churn Rate')

# Churn Rate by Charges Segment
sns.barplot(x=charges_churn_rate.index, y=charges_churn_rate.values, ax=axes[1])
axes[1].set_title('Churn Rate by Charges Segment')
axes[1].set_xlabel('Charges Segment')
axes[1].set_ylabel('Churn Rate')

# Churn Rate by Contract Type
sns.barplot(x=contract_churn_rate.index, y=contract_churn_rate.values, ax=axes[2])
axes[2].set_title('Churn Rate by Contract Type')
axes[2].set_xlabel('Contract Type')
axes[2].set_ylabel('Churn Rate')

plt.tight_layout()
plt.show()
```
